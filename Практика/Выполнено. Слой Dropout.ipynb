```python
import tensorflow as tf
```

```python
import numpy as np
from tensorflow.keras.layers import Dropout

# Входной тензор
X = np.ones((5, 2))

# Создание и вызов слоя в тестовом режиме
Dropout(0.2)(X)
```

```python
# Задание состояния
tf.random.set_seed(3)

# Создание и вызов слоя в тренировочном режиме
Dropout(0.2)(X, training=True)
```
